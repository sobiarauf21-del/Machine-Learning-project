```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
```

```python
X = df[['High','SP_close', 'DJ_close']]
Y = df['USO_Close']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
LR = LinearRegression()
LR.fit(X_train, Y_train)
Y_pred = LR.predict(X_test)
print(Y_pred)
```

```python
r2=r2_score(Y_test,Y_pred)
print('r2 Score:',r2)
```

```python
RMSE=root_mean_squared_error(Y_test,Y_pred)
print(RMSE)
```

```python
import joblib

# Suppose LR is your trained model
# Save the model
joblib.dump(LR, 'PROJECTML_model.pkl')
print("Model saved successfully!")

```

```python
import joblib
import pandas as pd
import gradio as gr


model = joblib.load('PROJECTML_model.pkl')



def predict_price(SP_close, DJ_close, High):
    
    input_data = pd.DataFrame([[SP_close, DJ_close, High]],  
                              columns=['SP_close', 'DJ_close', 'High'])
    
    prediction = model.predict(input_data)[0]
    
    return f"Predicted USO_Close Price: ${prediction:.2f}"


inputs = [
    gr.Number(label="S&P 500 Close Price (SP_close)"),
    gr.Number(label="Dow Jones Close Price (DJ_close)"),
    gr.Number(label="High Price (USO)")
]

interface = gr.Interface(
    fn=predict_price,
    inputs=inputs,
    outputs=gr.Textbox(label="Prediction Result"),
    title="USO Price Predictor (3 Key Features)",
    description="This model uses the most predictive variables: S&P Close, Dow Jones Close, and the High Price of USO."
)

interface.launch()
```
